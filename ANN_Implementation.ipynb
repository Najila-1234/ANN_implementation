```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# preprocessing steps
dataset = pd.read_csv('/content/breast_cancer.csv')
```

```python
dataset.head(10)
```

```python
# m  means breast cancer is there, b means no cancer
dataset.tail(10)
```

```python
dataset.shape
```

```python
x=dataset.iloc[:,2:32].values  # take full rows and colmns 2nd index colm to 32 colm
y=dataset.iloc[:,1].values  # output full rows and only first colm diagnosis
```

```python
x.shape  # 569 rows and 30 colmns
```

```python
y.shape   # 569 rows and 1 colm
```

```python
# do label encoding and replace m as 1 and b as 0
from sklearn.preprocessing import LabelEncoder
labelEncoder= LabelEncoder()
```

```python
y = labelEncoder.fit_transform(y)
```

```python
y
```

```python
# do train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)# shuffle colms randomly to increase accuracy
```

```python
# do scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
# start training ann
# create layers using keras
```

```python
import keras
from keras.models import Sequential  # seq for getting layers seqntially
from keras.layers import Dense   # dense for creating layers
```

```python
# initializing ANN
classifier = Sequential()
```

```python
# (nodes/neurons) units = (no of input features+1)/2
# units = (x+1)/2
# here x=30
# so (30+1)/2 = 15.5  = 16

# kernel_initializer is for assigning weights
# if we don't give input_dim then input layer is not there

# in 2nd hidden layer no need to define i/p layer becoz it is already defined, 1st hidden layer o/p
# is equal to 2nd hidden layer i/p
```

```python
# create first hidden layer and input layer
classifier.add(Dense(units = 16, kernel_initializer='uniform',activation='sigmoid',input_dim = 30))

# second hidden layer
classifier.add(Dense(units = 16, kernel_initializer='uniform',activation='sigmoid'))

# output layer
classifier.add(Dense(units=1,kernel_initializer='uniform', activation = 'relu'))
```

```python
# define all parameters like loss fn, optimizers for back propagation and forward propagation
# metrics to see output


classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(x_train,y_train,batch_size=10,epochs=100)
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred>0.5)
# takes only y_pred vlalues above 0.5
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```
